```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("spam.csv", encoding="latin-1")
```

```python
data.head()
```

```python
data.columns
```

```python
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data['class']=data['class'].map({'ham':0, 'spam':1})
```

```python
data.head()
```

```python
x=data['message']
y=data['class']
```

```python
x.shape
```

```python
y.shape
```

```python
data.isnull().sum()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
x_train_tfidf = vectorizer.fit_transform(x_train)
x_test_tfidf = vectorizer.transform(x_test)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()
lr.fit(x_train_tfidf, y_train)

pred_lr = lr.predict(x_train_tfidf)

print("Logistic Regression Accuracy:", accuracy_score(y_train, pred_lr))
```

#### SVM

```python
from sklearn.svm import LinearSVC

svm = LinearSVC()
svm.fit(x_train_tfidf, y_train)

pred_svm = svm.predict(x_train_tfidf)
print("SVM Accuracy:", accuracy_score(y_train, pred_svm))
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train_tfidf, y_train)

pred_rf = rf.predict(x_train_tfidf)
print("Random Forest Accuracy:", accuracy_score(y_train, pred_rf))
```

#### Naive Bayes 

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(x_train_tfidf, y_train)

pred_nb = nb.predict(x_train_tfidf)

print("Naive Bayes Accuracy:", accuracy_score(y_train, pred_nb))
```
